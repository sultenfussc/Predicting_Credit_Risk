```python
# Convert categorical data to numeric and separate target feature for testing data
y_train_dummies = pd.get_dummies(train_df['target'])
y_train = y_train_dummies[['high_risk']]
y_train
```

```python
# Convert categorical data to numeric and separate target feature for testing data
X = test_df.drop('target', axis = 1)
X_test_dummies = pd.get_dummies(X)

y_test_dummies = pd.get_dummies(test_df['target'])
y_test = y_test_dummies[['high_risk']]
y_test
```

```python
train_list = X_train_dummies.columns.to_list()
test_list = X_test_dummies.columns.to_list()
# print(train_list)


# check for differences between column lists
# print(list(set(train_list).symmetric_difference(set(test_list))))

print("missing from train_list: ",list(set(test_list) - set(train_list)))
print("missing from test_list: ", list(set(train_list) - set(test_list)))

print(f'train column count: {len(train_list)}')
print(f'test column count:  {len(test_list)}')
```

```python
# add missing dummy variables to testing set
for column in X_train_dummies.columns:
    if column not in X_test_dummies.columns:
        X_test_dummies[column] = 0
        
X_test_dummies
```

```python
train_list = X_train_dummies.columns.to_list()
test_list = X_test_dummies.columns.to_list()
# print(train_list)


# check for differences between column lists
# print(list(set(train_list).symmetric_difference(set(test_list))))

print("missing from train_list: ",list(set(test_list) - set(train_list)))
print("missing from test_list: ", list(set(train_list) - set(test_list)))

print(f'train column count: {len(train_list)}')
print(f'test column count:  {len(test_list)}')
```

### LogisticRegression and RandomForestClassifier Model Prediction:
I predict that the Random Forest Classifier will yield the most accurate model because it will produce many trees, or subgroups, that will encounter many different scenerios when processing the data.

```python
# Train the Logistic Regression model on the unscaled data and print the model score
classifier = LogisticRegression()
classifier

# Fit (train) our model by using the training data
classifier.fit(X_train_dummies,y_train)

#Validate the model by using the test data
print(f"Training Data Score: {classifier.score(X_train_dummies, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_dummies, y_test)}")
```

```python
# Create a confusion matrix for Logistic Regression
y_true = y_test
y_pred = classifier.predict(X_test_dummies)
confusion_matrix(y_true, y_pred)
```

```python
# The accuracy of the Logistic Regression model on the test data is TP + TN / (TP + FP + TN + FN)
tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
accuracy = (tp + tn) / (tp + fp + tn + fn) 
print(f"Accuracy: {accuracy}")
```

```python
print(classification_report(y_true, y_pred))
```

```python
# Train a Random Forest Classifier model and print the model score
clf = RandomForestClassifier(random_state=42, n_estimators=50).fit(X_train_dummies, y_train)

print(f'Training Score: {clf.score(X_train_dummies, y_train)}')
print(f'Testing Score: {clf.score(X_test_dummies, y_test)}')
```

### LogisticRegression and RandomForestClassifier Model Results:
The Random Forest Model performed better than the Logistic Regression Model. A thing to note is that the Random Forest Model is overfitting during the training data. This could raise concerns with how well it performed. But it is to be expected since the training data is already given information. Since, the testing data did not perform at the same level as the training data, we can feel more comfortable with the outcome. All in all, the returned outcomes reflect my initial prediction that the Random Forest would outperform the Logistic Regression Model.
